# Log in Data Science

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dict = {
    'company':['Tesla', 'Uber','Amazon','Lindal','Naver','Vedanta', 'Axis Bank'],
    'revenue':[31.0, 11.0, 386.0, 4.7, 5.6, 11.3,8.2]
}

df = pd.DataFrame(df_dict)
df
```

```python
plt.bar(df['company'], df['revenue'])
plt.legend(['revenue'], loc ='upper right' )
plt.show()
```

Because Amazon's revenue is too high, that's why it is difficult to compare other companies' revenue. It seems like they're almost similar.    
That's why we can use 'log' 

```python
df.plot(x = 'company', y='revenue', kind = 'bar' ,logy = True)
plt.legend(['revenue'], loc ='upper right' )
plt.ylabel('revenue in log format')
plt.show()
```

range: 0 ~ 10^1 == revenue is between 0 ~ 100      
range: 10^1 ~ 10^2  == revenue is btw 100 ~ 1000


> By using LOG we can get better comparizon when some value is too large and other is in small scale
